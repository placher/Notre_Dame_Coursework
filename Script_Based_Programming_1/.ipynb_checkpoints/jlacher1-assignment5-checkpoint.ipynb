#### Return a list of all samples that occur once (hapax legomena)  Return type:	list

#### fourLetterWords( ) takes no arguements and returns a list with each element being a four-letter word in text2

```python
def fourLetterWords():
    """Returns a list with each element being a four-letter word in text2"""
    mylist = ['']
    text2lower = [w.lower() for w in text2]
    for word in text2lower:
        if len(word) == 4:
            mylist.append(word)
    return mylist
```

```python
fourletterwords = fourLetterWords()
```

#### flw(x) is an improved version of fourLetterWords() that takes a text file as an argument and returns a lowercase list of four letter words.

```python
def flw(x):
    """Takes a text file as an argument and returns a lowercase list of four letter words."""
    y = [w.lower() for w in x if len(w) == 4]
    return y
```

#### nlw(x,n) takes a text file and a number as an argument and returns a lowercased set of words of length n.

```python
def nlw(x,n):
    """Takes a text file and an interger as arguments and returns returns a lowercased set of words of length n"""
    z = [w.lower() for w in x if len(w) == n]
    return z
```

```python
print['n=2: ',len(nlw(text2,2))]
print['n=4: ',len(nlw(text2,4))]
print['n=8: ',len(nlw(text2,8))]
print['n=12: ',len(nlw(text2,12))]
```
