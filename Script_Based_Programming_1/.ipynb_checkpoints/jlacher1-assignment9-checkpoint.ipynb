##Task 2: Flip, Flop and Fly##


In this task, you will create three copies of your original meme image. One will be flipped vertically (so that what was the bottom is now the top and _vice versa_).  One will be flipped horizontally (left for right), and one will be rotated by 90 degrees counterclockwise (so that the former right side is the new top side, _etc_.).

1. Create a vertically flipped image from your original image and display it.

2. Same as 1, but horizontally flipped.

3. Same as 1, but rotated counterclockwise by 90 degrees.

__Note__: the `im.transpose()` method in PIL may be of use. 

````python
# your code goes here vertical flip
im2 = im.transpose(Image.FLIP_TOP_BOTTOM)
plt.imshow(im2)
plt.axis('off')
````

````python
# your code goes here horizontal flip
im3 = im.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(im3)
plt.axis('off')
````

````python
# your code goes here 90 degree rotation
im4 = im.transpose(Image.ROTATE_90)
plt.imshow(im4)
plt.axis('off')
````

## Task 3: Captions I ##

Let's put some text on the 90-degrees-rotated version of the image.  We will use the PIL ImageFont and ImageDraw classes to do this.

+ Think up a clever and G-rated caption for your image.
+ Use the following code to load a fairly generic font into Python. The integer "40" means that the letters of the font are about 40 pixels tall (more or less).

````python
# make font directory, if necessary
if not os.path.exists('fonts'):
    os.makedirs('fonts')
# retrieve font file, if necessary
if not os.path.exists('fonts/Geneva.ttf'):
    urllib.urlretrieve('http://patrick-flynn.github.io/code/meme/fonts/Geneva.ttf','fonts/Geneva.ttf')
# load font
geneva = ImageFont.truetype('fonts/Geneva.ttf',40)
````

````python
# your code goes here create ImageDraw object

cap1 = ImageDraw.Draw(im4,'RGBA')
````

+ Draw your motto using the ImageDraw `text()` method, at a position of (0,0), using the Geneva font you loaded earlier.  Display your results.

The (0,0) position is the upper-left corner of the image.

````python
# your code goes here, caption upper left corner of 90 degree rotated image
cap1.text((0,0),'Why Does It Never Work The First Time',font=geneva,fill=(255,255,255))
plt.imshow(im4)
plt.axis('off')
````

You may choose a method for drawing the caption so that it is always easy to read. One possibility is to draw a filled rectangle in the proper place. Display your results.

````python
# your code goes here, centered contrasty caption

# make font directory, if necessary
if not os.path.exists('fonts'):
    os.makedirs('fonts')
# retrieve font file, if necessary
if not os.path.exists('fonts/Courier.ttf'):
    urllib.urlretrieve('http://patrick-flynn.github.io/code/meme/fonts/Courier.ttf','fonts/Courier.ttf')
# load font
Courier = ImageFont.truetype('fonts/Courier.ttf',60)

# make font directory, if necessary
if not os.path.exists('fonts'):
    os.makedirs('fonts')
# retrieve font file, if necessary
if not os.path.exists('fonts/CourierBold.ttf'):
    urllib.urlretrieve('http://patrick-flynn.github.io/code/meme/fonts/CourierBold.ttf','fonts/CourierBold.ttf')
# load font
CourierBold = ImageFont.truetype('fonts/CourierBold.ttf',60)

txt = 'Why Does It Never Work The First Time'
txtorigin = (im.size[0] - CourierBold.getsize(txt)[0])/2.
cap2 = ImageDraw.Draw(im,'RGBA')
cap2.text((txtorigin,0),txt,font=CourierBold,fill=(255,255,255))
plt.imshow(im)
plt.axis('off')
````
